```python
import numpy
import bokeh.plotting
```

```python
# Set the output to 'notebook' mode
bokeh.plotting.output_notebook()
```

```python
# Simple numpy calculation of a sine wave
x = numpy.arange(0.0, 100.0, 0.1)
y = numpy.sin(x)
```

```python
p = bokeh.plotting.figure(title="Simple line example", x_axis_label='x', y_axis_label='y')
p.line(x,y, line_width=3.0)
bokeh.plotting.show(p)
```
